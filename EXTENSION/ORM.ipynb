```python
from sklearn.ensemble import IsolationForest as IF
```

```python
X,Y = getBFS(part=[0,-1],dataset='a')
ALL=np.concatenate([X.values, Y.reshape(-1,1)],axis=1)
clf = IF(contamination=0.1)
clf.fit(ALL)
OUT = clf.predict(ALL)
```

```python
#plt.title('Arterial roads data set with 10% points found to be outliers using Isolation Forest')
plt.figure(figsize=(5.5,4.5))
plt.hist(np.log1p(Y),bins=30, label='All data')
plt.hist(np.log1p(Y[OUT==-1]),bins=30,label='10% outliers (IF)')
plt.xticks(np.arange(1,8),np.round(np.expm1(np.arange(1,8))).astype(int))
plt.xlabel('Incident duration [min]', fontsize=16)
plt.ylabel('Number of instances [n]', fontsize=16)
plt.text(6.1, 65, 'a)', fontsize = 23)
plt.tight_layout()
plt.legend(fontsize=12)
plt.savefig("ORM_A.pdf")
print()
```
