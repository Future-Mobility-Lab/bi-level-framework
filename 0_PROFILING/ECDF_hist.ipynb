```python
import numpy as np
def ecdf(data):
    """ Compute ECDF """
    x = np.sort(data)
    n = x.size
    y = np.arange(1, n+1) / n
    return(x,y)
A,B=ecdf(Y)
plt.plot(A,B)

plt.axvline(40,ls='--',c='r',label='(40 min)')
plt.axvline(60,ls='--',c='r',label='(60 min)')
plt.axvline(125,ls='--',c='r',label='(125 min)')
# plt.axvline(np.median(Y),ls='-',c='r',label='Median')
plt.ylim(0,1)

plt.minorticks_on()
plt.ylabel('Probability (T < Duration)')
plt.xlabel('Incident duration [min]')
plt.grid(b=True,which='both')
plt.legend()
plt.savefig('aecdf.eps',format='eps')
```
